```python
import pandas as pd
import numpy as np
```

```python
data=pd.read_csv("titanic.csv")
data.head(50)
```

```python
data.info()
```

```python
data.columns
```

```python
data.columns=['passengerid', 'survived', 'pclass', 'lastname', 'name', 'sex', 'age',
       'sibsp', 'parch', 'ticket', 'fare', 'cabin', 'embarked']
```

```python
data.info()
```

```python
data.isna().sum()
```

```python
# median method to impute missing value 
data["age"]
```

```python
age_median = data["age"].median()
age_median
```

```python
data["age"]=data["age"].fillna(age_median)
```

```python
data.isna().sum()
```

```python
data.drop(["cabin"],axis=1,inplace=True)
```

```python
data.info()
```

```python
import warnings
warnings.filterwarnings("ignore")
```

```python
embarked_mode=data["embarked"].value_counts()[1]
embarked_mode
```

```python
data["embarked"]=data["embarked"].fillna(embarked_mode)
```

```python
data.isna().sum()
```

```python
clean_data=data.drop(["survived","lastname","name","ticket","passengerid"],axis=1)
```

```python
clean_data["family_size"]=clean_data["sibsp"] + clean_data["parch"]+ 1
```

```python
clean_data=clean_data.drop(["sibsp","parch"],axis=1)
```

```python
dummy_variables=pd.get_dummies(clean_data,drop_first=True,dtype=int)
```

```python
dummy_variables
```

```python
x= dummy_variables
x
```

```python
y=data["survived"]
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=243,shuffle=True)
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
model=LogisticRegression()
model
```

```python
model.fit(x_train,y_train)
```

```python
y_pred=model.predict(x_test)
```

```python
x_test.columns
```

```python
model.predict([[2,34,5000,4,0,1,0,0]])
```

```python
from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report
```

```python
confusion_matrix(y_pred,y_test)
```

```python
pd.DataFrame(data=confusion_matrix(y_pred,y_test),columns=["Actual Negative","ActualPositive"],index=["Predicted Negative","Predicted Positive"])
```

```python
from sklearn.metrics import accuracy_score,f1_score,recall_score,precision_score
```

```python
accuracy_score(y_pred,y_test)
```

```python
f1_score(y_pred,y_test)
```

```python
recall_score(y_pred,y_test)
```

```python
precision_score(y_pred,y_test)
```

```python
print(classification_report(y_pred,y_test))
```
